1) What is the difference between interpreted and compiled languages?
 - Difference between interpreted and compiled languages:
 -Interpreted languages (like Python) are executed line-by-line by an interpreter. The interpreter converts the high-level code into machine code at runtime, which can slow down performance.
 -Compiled languages (like C, C++) are transformed into machine code ahead of time by a compiler. This machine code is then executed directly by the system, leading to faster execution.

2) What is exception handling in Python?
 -  Exception handling in Python is used to manage runtime errors or unexpected situations. It uses try, except, else, and finally blocks to ensure that the program can continue running even when an error occurs.

3) What is the purpose of the finally block in exception handling?
 - The finally block is executed no matter what, even if an exception is raised or not. It's commonly used for cleanup activities such as closing files or releasing system resources.

4) What is logging in Python?
 - Logging in Python allows you to record messages about your program's execution. These messages can be useful for debugging, monitoring, and auditing. Python's logging module provides a flexible way to capture logs at various levels (e.g., DEBUG, INFO, WARNING).

5) What is the significance of the __del__ method in Python?
 - The __del__ method is a special method that is called when an object is about to be destroyed. It’s used to perform any necessary cleanup, such as closing files or releasing other resources.

6) What is the difference between import and from ... import in Python?
 - Difference between import and from ... import in Python:
 - import module_name imports the entire module, and you access its functions via module_name.function().
 - from module_name import function_name imports a specific function or class directly, allowing you to call function_name() without the module prefix.

 - try:

except (ExceptionType1, ExceptionType2) as e:
    # handle exceptions

8) What is the purpose of the with statement when handling files in Python?

 - with open("file.txt", "r") as file:
    # process file

9) What is the difference between multithreading and multiprocessing?
 - Difference between multithreading and multiprocessing:
 - Multithreading: Multiple threads within a single process share the same memory space. It is suitable for I/O-bound tasks but is limited by Python’s Global Interpreter Lock (GIL) for CPU-bound tasks.
 - Multiprocessing: Involves multiple processes running concurrently, each with its own memory space. This is suitable for CPU-bound tasks as it can fully utilize multiple processors.

10) What are the advantages of using logging in a program?
 - Logging helps in tracking program execution, identifying and diagnosing errors, recording events for audit purposes, and debugging issues in production environments. It also provides a way to monitor long-running applications.

11) What is memory management in Python?
 - Python uses automatic memory management, including reference counting and garbage collection, to handle memory allocation and deallocation. The garbage collector reclaims memory from objects that are no longer in use.

12) What are the basic steps involved in exception handling in Python?
 - Basic steps involved in exception handling in Python:-
 - Try: Enclose code that may raise an exception.
 - Except: Catch specific exceptions and handle them.
 - Else: Execute code when no exception occurs.
 - Finally: Always execute code, used for cleanup (like closing files).

13) Why is memory management important in Python?
 - Proper memory management is essential for ensuring that resources (like memory and file handles) are used efficiently and are freed when no longer needed. This helps prevent memory leaks and improves performance.

14) What is the role of try and except in exception handling?
 - Role of try and except in exception handling:
 - try: Defines a block of code that might cause an exception.
 - except: Catches the exception and allows you to handle it, rather than crashing the program.

15) How does Python's garbage collection system work?
 - Python's garbage collection uses reference counting to track the number of references to an object. When no references remain, the object is deleted. Additionally, Python cyclic garbage collector handles circular references by periodically checking for unreachable objects.

16) What is the purpose of the else block in exception handling?
 - Purpose of the else block in exception handling: The else block is executed if no exceptions were raised in the try block. It's useful for code that should only run when no errors occur.

17) What are the common logging levels in Python?
 - Common logging levels in Python:
 - DEBUG: Detailed information for debugging.
 - INFO: General information about the program’s execution.
 - WARNING: Indicates a potential problem.
 - ERROR: An issue that prevents some functionality.
 - CRITICAL: A serious error that may terminate the program.

18) What is the difference between os.fork() and multiprocessing in Python?
 - Difference between os.fork() and multiprocessing in Python:
 - os.fork(): A system call in Unix-like OS that creates a new process by duplicating the parent process. It has limited cross-platform support.
 - multiprocessing: A Python module that creates separate processes with independent memory spaces, suitable for cross-platform parallelism and CPU-bound tasks.

19) What is the importance of closing a file in Python?
 - Closing a file ensures that changes are saved and that system resources (like file handles) are released. Not closing files can lead to resource leaks and potential file corruption.

22) What is the os module in Python used for in file handling?

23) What are the challenges associated with memory management in Python?

16. Demonstrate how to use memory profiling to check the memory usage of a small program.

```python
import psutil
import os

# Function to demonstrate memory usage
def memory_usage():
    process = psutil.Process(os.getpid())
    print(f"Memory usage: {process.memory_info().rss / 1024 / 1024} MB")

# Example function to check memory usage
def my_function():
    memory_usage()  # Check memory usage at the start
    a = [1] * (10**6)  # List of 1 million integers
    memory_usage()  # Check memory usage after creating the list
    b = [2] * (2 * 10**7)  # List of 20 million integers
    memory_usage()  # Check memory usage after creating the second list
    del b  # Delete the second list to free memory
    memory_usage()  # Check memory usage after deleting the list
    return a

if __name__ == "__main__":
    my_function()
```

18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?

```python
import logging
from logging.handlers import RotatingFileHandler

# Create a rotating file handler
handler = RotatingFileHandler('app.log', maxBytes=1e6, backupCount=5)
logging.basicConfig(level=logging.INFO, handlers=[handler])

logging.info("This is a log message.")
```

21. Write a Python program that reads a file and prints the number of occurrences of a specific word.


```python
word_to_find = "Python"
count = 0

with open("file.txt", "r") as file:
    for line in file:
        count += line.lower().count(word_to_find.lower())

print(f"The word '{word_to_find}' occurred {count} times.")
```
